```python
import numpy as np
import pandas as pd
from pyensembl import EnsemblRelease


ens = EnsemblRelease(99)
```

```python
known_genes = pd.read_csv('/s/project/gtex-processed/als/data/processed/literature/nature_review_als_genes.csv')
known_genes.head()
```

```python
gene_name_mapping = pd.read_csv('../graphgenefinder/hugo_symbol_to_ENSG_mapping.tsv', sep = '\t')
gene_name_mapping.head()
```

```python
known_genes = known_genes.merge(gene_name_mapping, left_on='gene_name', right_on='any_symbol')
known_genes
```

```python
unique_known_genes = np.unique(known_genes.Ensembl_gene_ID)
```

```python
len(unique_known_genes)
```

```python
import geneprof.read_databases 
from geneprof.graph_features import RandomWalk
```

```python
%%time
graphs = geneprof.read_databases.read_data()
```

```python
rw = RandomWalk(graphs['string'])
```

```python
genes = pd.DataFrame({'gene_id': graphs['string'].nodes()})
```

```python
res = rw.get_feature(unique_known_genes, genes)
```
